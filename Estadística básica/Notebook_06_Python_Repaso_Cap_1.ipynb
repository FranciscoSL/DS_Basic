## Repaso conceptual

Las siguientes son ideas, conceptos y referencias a materiales relacionados con conceptos introductorios al estudio de Ciencia de Datos.

+ Ciencia de datos integra conocimientos de múltiples ciencias y disciplinas, entre ellas: matemáticas y estadística, y ciencias de la computación. Estos conocimientos se combinan con el expertise en una industria o negocio particular.
+ Dentro de los lenguajes más usados para Ciencia de datos se destaca [Python](https://www.crehana.com/co/blog/desarrollo-web/que-es-python/) y [R](https://www.datanalytics.com/libro_r/programacion-en-r.html). Asu ves, dentro de cada lenguaje existen múltiples [librerías](https://relopezbriega.github.io/blog/2014/05/28/python-librerias-esenciales-para-el-analisis-de-datos/) (o paquetes) con funciones y herramientas que facilitan el análisis y procesamiento de los datos.
+ En el análisis de datos hay diversos conceptos cuando se habla de [tipos de variables](https://support.minitab.com/es-mx/minitab/18/help-and-how-to/modeling-statistics/regression/supporting-topics/basics/what-are-categorical-discrete-and-continuous-variables/). Se destacan estas clasificaciones:<br/>
    * Variables categóricas<br/>
    * Variables cuantitativas (discretas, continuas)<br/>
<br/>
+ Cuando la Ciencia de datos se aplica a grandes volúmenes de datos, se refiere a aplicaciones a Big Data. Un  concepto relevante de Big Data está relacionado con las categorías o tipos de datos según su formato. En estas aplicaciones se puede encontrar conjuntos de datos estructurados, no estructurados o semiestruturados.
+ Dada la demanda técnica de recursos (máquinas) en el procesamiento de grandes volúmenes de datos, el cloud computing se utiliza masivamente para soportar aplicaciones robustas de Big Data. Cloud computing no solo responde a la demanda técnica, si no que genera beneficios adicionales para la gestión de los datos y la información.

## Repaso práctico

Si bien para cualquier proceso de análisis de datos es estandar realizar iniciar con Análisis Exploratorio de Datos (EDA), Transformación de datos e Ingeniería de features, para efectos de repaso de las herramientas del curso, se presenta únicamente una exploración preliminar del dataset cars.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = pd.read_csv("cars.csv")
print()
print('El dataset tiene ' + str(df.shape[0]) + ' filas de datos y ' + str(df.shape[1]) + ' columnas de datos.')
print()
print('Esta es una muestra (primeras líneas) de la estructura del dataset:')
print()
df.head()
```

Los nombres de las columnas y su tipo de dato son:

```python
#df.info()
#df.astype
df.dtypes
```

Diccionario de datos:

Voy a exceptuar las columnas feature 0 a la 9 de este análisis. Entonces las elimino del DataFrame.

```python
df.drop(columns = ['feature_0','feature_1','feature_2', 'feature_3','feature_4','feature_5','feature_6', 'feature_7','feature_8','feature_9'], inplace = True)
print(df.shape)
df.head()
```

Reconocimiento de algunas variables:

```python
#Cuántas manufacturers hay?
print(df.manufacturer_name.count())
print(df.manufacturer_name.value_counts()) # ordenados
print()
plt.figure(figsize = (10,6))
plt.xticks(rotation = 45)
#sns.countplot(data = df, x = "manufacturer_name", linewidth=1,edgecolor=(0,0,0),palette="Set2")
df.manufacturer_name.value_counts().plot(kind='bar', alpha=0.4, color='coral', linewidth=1, edgecolor=(0,0,0))
plt.title('Vehículos por manufacturer')
plt.show()
```

```python
df.groupby('manufacturer_name').count()
```

```python
kia.groupby('body_type').mean()
kia.groupby('body_type')['price_usd'].mean() # Este es el precio promedio por body_type dentro de Kia
```

```python
#Esto también se puede hacer de la siguiente forma:
for key_group, group in kia.groupby('body_type'):
    grouped_price = group['price_usd'].mean()
    
    print('body_type:{}, Price: {}\n'.format(key_group,grouped_price))
```
