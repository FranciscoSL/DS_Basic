### Pequeño repaso de listas

<font color='darkcyan'>**Ejercicio:**</font> dadas las siguientes listas a y b
```python
a = [12,3,5,12,8,9,25,12,1,3,4,5,17,15]
b = [3,5,7,8,45,8,9,16,1,11,30,41,2,3,15,1,29,22]
```

Armar una nueva lista `c` con:

1. Con todos los elementos de *a* y *b*
1. Con todos los elementos de *a* y *b*, ordenados de menor a mayor
1. Con todos los elementos de *a* y *b*, ordenados de mayor a menor
1. Con todos los elementos de *a* y *b*, ordenados de menor a mayor y sin elementos repetidos
1. Con todos los elementos que están en ambas listas, sin elementos repetidos. Es decir, si un elemento está sólo en una lista, no debe aparecer en `c`.
1. Con todos los elementos que NO están en ambas listas, sin elementos repetidos. Es decir, si un elemento está en `a` y en `b`, no debe aparecer en `c`.

**Pistas:** es posible resolver este ejercicio trabajando con listas, arrays o sets. Según el ítem de este ejercicio, una estructura de datos puede ser más apropiada que las demás.

````python
a = [12,3,5,12,8,9,25,12,1,3,4,5,17,15]
b = [3,5,7,8,45,8,9,16,1,11,30,41,2,3,15,1,29,22]
````

````python
lista_c1 = a + b
print('lista_c1 =',lista_c1)
lista_c2 = sorted(a + b)
print('lista_c2 =',lista_c2)
lista_c3 = sorted(a + b, reverse = True)
print('lista_c3 =',lista_c3)
lista_c4 = list(set(sorted(a + b)))
print('lista_c4 =',lista_c4)

lista_5a = list(set(sorted(a)))
lista_5b = list(set(sorted(b)))
lista_5ab = list(set(sorted(lista_5a + lista_5b)))

lista_c5 = []

for i in lista_5ab:
    if i in lista_5a and i in lista_5b:
        lista_c5.append(i)
print('lista_c5 =',lista_c5)

lista_c6 = []

for i in lista_5ab:
    if i not in lista_c5:
        lista_c6.append(i)
print('lista_c6 =',lista_c6)
````

Hay dos formas de utilizar Matplotlib. Se recomienda ver este [tutorial](https://matplotlib.org/3.1.3/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py).
- Una forma es a través de los comandos `plt.`. Esta forma es más rápida y sencilla, pero menos versátil.

### 1. Gráficos de Líneas y Puntos
Vamos a graficar una función seno, de tipo: $y = sen(2 \pi x) + 1$.

La manera mas rápida de lograr un gráfico es la siguiente:

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.arange(0.0, 3.0, 0.05)  # Sirve para definir los valores del eje X
y = 1 + np.sin(2 * np.pi * x) # Se genera la función y(x)

# Simplemente le pedimos a la librería que dibuje lo que previamente se ha definido
plt.plot(x, y)
plt.title('Gráfico (de línea) de una función Seno'); # Se define el título del gráfico. El ; evita que salga la línea de código en el gráfico
# plt.show() # Comando opcional en este caso, pero requerido en otros.
````

Si queremos pasar este gráfico a uno de puntos, es muy sencillo:

````python
plt.plot(x, y, 'o')
plt.title('Gráfico (de puntos) de una función Seno')
plt.show()
````

O de puntos y líneas:

````python
plt.plot(x, y, 'o-')
plt.title('Gráfico (puntos y líneas) de una función Seno')
plt.show()
````

Más opciones de Markers en la [documentación](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

Muy parecida es la función `plt.scatter()`, que sirve para hacer gráficos de dispersión. A diferencia de `plt.plot()`, esta función no permite gráficos de líneas.

````python
plt.scatter(x, y)
plt.title('Gráfico de dispersión (solo puntos)')
plt.show()
````

### 2. Agregando información al gráfico

Por favor asegurarse siempre que su gráfico sea **entendible** y **claro** para **cualquier persona**. Agregar información a los gráficos sin saturarlos favorece su comprensión. Algunos adicionales importantes:

* Etiquetar correctamente los ejes. También usar unidades
* Agregar título
* En ocasiones agregar etiquetas de datos

````python
x = np.arange(0.0, 24, 0.5)
y = 25 + 6*np.sin(2*np.pi*(x - 8)/24)

plt.plot(x, y, color = 'g', linewidth = 2)

# Info adicional para el gráfico
plt.xlabel('Tiempo (horas)')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura de un día en Medellín')

plt.grid() # Este comando activa líneas de cuadrícula.
plt.show() #Show muestra el gráfico. Si no lo ponemos, en general no pasa nada. Pero a veces es necesario.
````

**Leyendas**

En gráficos con múltiples curvas se usan leyendas para distinguir las curvas.

````python
x = np.arange(0.0, 24, 0.5)
y1 = 25 + 6*np.sin(2*np.pi*(x - 8)/24)
y2 = 24 + 6*np.sin(2*np.pi*(x - 8)/24)

plt.plot(x, y1, label = 'Regular')
plt.plot(x, y2, label = 'Época de invierno')

plt.xlabel('Tiempo (horas)')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura diaria promedio')

plt.grid()

# Se agregan leyendas al gráfico
plt.legend() # Editar la leyenda: loc='upper center', shadow=True, fontsize='x-large')

plt.show() 
````

### 3. Histogramas

Matplotlib también permite graficar histogramas mediante la función `hist`.

````python
# Elegimos una distribución con centro en 20 y una desviación igual a 4
mu = 20  # Media de la distribución
sigma = 4  # Desviación Estandar
valores = mu + sigma * np.random.randn(1000)

# Imprimimos la media y la desviación de los valores generados
print('Media:',np.mean(valores))
print('Desv. std.:',np.std(valores),'\n')
# Tener presente que mientas más valores se seleccionen, más se aproximarán a la distribución normal.

# Ahora se grafica el histograma para estos valores
plt.hist(valores, bins = 50)
plt.show()
````
